```python
import matplotlib.pyplot as plt

# Your data
age = [0, 0, 2, 11, 7, 16, 9, 0, 12, 4]
baseline_nox_emissions = [1.71, 4.35, 4.08, 1.28, 5.32, 0.57, 3.39, 3.41, 0.75, 1.24]

# Scatter plot
plt.scatter(age, baseline_nox_emissions, color='blue', label='Baseline NOx Emissions')
plt.title('Baseline NOx Emissions vs Age')
plt.xlabel('Age')
plt.ylabel('Baseline NOx Emissions')
plt.legend()
plt.show()
```
